```python
import numpy as np
import pandas as pd
import os

from datasets import load_dataset, load_metric

from bert_score import BERTScorer
from rouge import Rouge

%matplotlib inline
```

```python
def add_rouge(df):
    rouge = Rouge()
    result = rouge.get_scores(df['Generated Text'], df['Actual Text'])
    df["Rouge-1"] = [a_dict['rouge-1']['f'] for a_dict in result]
    df["Rouge-2"] = [a_dict['rouge-2']['f'] for a_dict in result]
    df["Rouge-l"] = [a_dict['rouge-l']['f'] for a_dict in result]
    return df
```

```python
path = 'output'
total_score = pd.DataFrame()
for file in os.listdir(path):
    df = pd.read_csv(path+"/"+file)
    df = add_rouge(df)
    total_score = pd.concat([total_score, df])
    print(file)
total_score
```

```python
total_score.reset_index(drop=True, inplace=True)
total_score
```

```python
total_score.to_csv("Results/Total_score.csv")
```
