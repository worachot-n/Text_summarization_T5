```python
import numpy as np
import pandas as pd
import os

from datasets import load_dataset, load_metric

from bert_score import BERTScorer
from rouge import Rouge

%matplotlib inline
```

```python
def add_score(df):
    # Rouge score
    rouge = Rouge()
    result = rouge.get_scores(df['Generated Text'], df['Actual Text'])
    df["Rouge-1"] = [a_dict['rouge-1']['f'] for a_dict in result]
    df["Rouge-2"] = [a_dict['rouge-2']['f'] for a_dict in result]
    df["Rouge-l"] = [a_dict['rouge-l']['f'] for a_dict in result]
    # BERT score
    scorer = BERTScorer(lang="en", rescale_with_baseline=True)
    refs = df['Actual Text'].to_list()
    hyps = df['Generated Text'].to_list()
    P, R, F1 = scorer.score(hyps, refs)
    df["BERTscore"] = F1
    return df
```

```python
Real_df = pd.read_csv("Real_Evaluate_Summarize.csv")
path = 'output'
total_score = pd.DataFrame()
df_cat = pd.DataFrame()
for _, file in enumerate(os.listdir(path)):
    df = pd.read_csv(path+"/"+file)
    df_cat = pd.concat([df_cat, df])
    print(_, file)
    if _%6 == 5:
        df_cat["Actual Text"] = Real_df["Real summarize"]
        df_cat = add_score(df_cat)
        total_score = pd.concat([total_score, df_cat])
        df_cat = pd.DataFrame()
total_score
```
